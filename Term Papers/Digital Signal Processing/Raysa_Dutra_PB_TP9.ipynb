### Questão 5

5. Simular, por soma de senos e cossenos, o sinal de áudio, simulando o timbre, ou variar a frequência dos áudios apresentados no primeiro item.


```
# Com base no item anterior, vou utilizar a frequência dada para cada instrumento para fazer uma soma de cada:

# Seno
# Flauta
sin_flute_signal = thinkdsp.SinSignal(freq=830.61, amp=1.0, offset=0)
sin_flute_signal.plot()
```

```
# Cos
# Flauta
cos_flute_signal = thinkdsp.CosSignal(freq=801.82, amp=1.0, offset=0)
cos_flute_signal.plot()
```

```
# Soma
# Flauta
sum_flute_signal = cos_flute_signal + sin_flute_signal
sum_flute_signal.plot()
```

```
# Seno
# Violino
sin_violin_signal = thinkdsp.SinSignal(freq=1567.99, amp=1.0, offset=0)
sin_violin_signal.plot()
```

```
# Cos
# Violino
cos_violin_signal = thinkdsp.CosSignal(freq=440, amp=1.0, offset=0)
cos_violin_signal.plot()
```

```
# Soma
# Violino
sum_violin_signal = sin_violin_signal + cos_violin_signal
sum_violin_signal.plot()
```


### Questão 6

6. Sintetizar a escala musical a partir da nota simulada na questão 5.



```
# Reprodução de áudio da soma da flauta apresentada na 5 que foi feita com base no segmento:
audio_sum_flute_signal = sum_flute_signal.make_wave(duration=0.01, start=0, framerate=11025)
audio_sum_flute_signal.normalize()
audio_sum_flute_signal.make_audio()
```

```
# Áudio original do Segmento
segment_flute.make_audio()
```

```
# Reprodução de áudio da soma do violino apresentada na 5 que foi feita com base no segmento
audio_sum_violin_signal = sum_violin_signal.make_wave(duration=0.01, start=0, framerate=11025)
audio_sum_violin_signal.normalize()
audio_sum_violin_signal.make_audio()
```

```
# Áudio original do Segmento
segment_violin.make_audio()
```

```
# Sintetização da escala musical utilizando os dados do áudio original:
def synthesize(amps, fs, ts):
    components = [thinkdsp.CosSignal(freq, amp)
                  for amp, freq in zip(amps, fs)]
    signal = thinkdsp.SumSignal(*components)

    ys = signal.evaluate(ts)
    return ys
```

```
# Flauta

amps_flute = np.array([1.0, 48.20])
fs_flute = [2405.45, 801.82]
framerate = 11025

ts = np.linspace(0, 1, framerate)
flute_ys = synthesize(amps_flute, fs_flute, ts)
new_flute_wave = thinkdsp.Wave(flute_ys, framerate)
new_flute_wave.normalize()
new_flute_audio = new_flute_wave.make_audio()
new_flute_audio
```

```
# Violino

amps_violin = np.array([1.0, 53.01])
fs_violin = [1900.00, 400.00]
framerate = 11025

ts = np.linspace(0, 1, framerate)
violin_ys = synthesize(amps_violin, fs_violin, ts)
new_violin_wave = thinkdsp.Wave(violin_ys, framerate)
new_violin_wave.normalize()
new_violin_audio = new_violin_wave.make_audio()
new_violin_audio
```
